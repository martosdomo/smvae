```python
from smvae.c_mnist import *
from smvae.models import *
from smvae.functions import *

from google.colab import files
```

```python
# HYPERPARAMETERS

LATENT_SIZE = 10
BATCH_SIZE = 32
LEARNING_RATE = 0.5*1e-3
NUM_EPOCHS = 15

random_seed = 2
is_bce = False
```

```python
random.seed(random_seed)

digit_inst = 4
length = 24000
min_contrast = 0
max_contrast = 1

trainset = create_trainset(digit_inst, length, min_contrast, max_contrast)
testset = create_testset(digit_inst)


testloader = DataLoader(
    testset,
    batch_size=1,
    shuffle=True
)
```

```python
torch.manual_seed(random_seed)
torch.cuda.manual_seed(random_seed)
torch.backends.cudnn.deterministic = True
torch.backends.cudnn.benchmark = False

enc_layers = [256, 32]
dec_layers = [32, 256]

model_vae2 = VAE(784, enc_layers, dec_layers, LATENT_SIZE, 2, nn.ReLU(), nn.ReLU())
model_vae1 = VAE(784, enc_layers, dec_layers, LATENT_SIZE, 1, nn.ReLU(), nn.ReLU())
model_vae0_5 = VAE(784, enc_layers, dec_layers, LATENT_SIZE, 0.5, nn.ReLU(), nn.ReLU())
model_vae0_2 = VAE(784, enc_layers, dec_layers, LATENT_SIZE, 0.2, nn.ReLU(), nn.ReLU())
model_vae0_1 = VAE(784, enc_layers, dec_layers, LATENT_SIZE, 0.1, nn.ReLU(), nn.ReLU())
model_vae0_05 = VAE(784, enc_layers, dec_layers, LATENT_SIZE, 0.05, nn.ReLU(), nn.ReLU())

model_lognormal2 = SMVAE_LOGNORMAL(784, enc_layers, dec_layers, LATENT_SIZE, 2, nn.ReLU(), nn.ReLU())
model_lognormal1 = SMVAE_LOGNORMAL(784, enc_layers, dec_layers, LATENT_SIZE, 1, nn.ReLU(), nn.ReLU())
model_lognormal0_5 = SMVAE_LOGNORMAL(784, enc_layers, dec_layers, LATENT_SIZE, 0.5, nn.ReLU(), nn.ReLU())
model_lognormal0_2 = SMVAE_LOGNORMAL(784, enc_layers, dec_layers, LATENT_SIZE, 0.2, nn.ReLU(), nn.ReLU())
model_lognormal0_1 = SMVAE_LOGNORMAL(784, enc_layers, dec_layers, LATENT_SIZE, 0.1, nn.ReLU(), nn.ReLU())
model_lognormal0_05 = SMVAE_LOGNORMAL(784, enc_layers, dec_layers, LATENT_SIZE, 0.05, nn.ReLU(), nn.ReLU())

model_gamma = SMVAE_GAMMA(784, enc_layers, dec_layers, LATENT_SIZE, alpha = 1)
```

```python
#models = [model_vae, model_lognormal, model_gamma]
#models = [model_vae2, model_vae1, model_vae0_5, model_vae0_2, model_vae0_1, model_vae0_05]
models = [model_lognormal1, model_lognormal0_5, model_lognormal0_2, model_lognormal0_1, model_lognormal0_05]

NUM_EPOCHS = 10

for model in models:
    myloss = train(model, trainset, is_bce, LEARNING_RATE, BATCH_SIZE, NUM_EPOCHS)
    plot(myloss, model.name)
```
